## Function : Create project catalog

```python
# Function : create project catalog
def catalog_setup(catalog):
    setup_res = "OK"

    try:
        spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog}")
    except Exception as e:
        print("Error : ", e)
        setup_res = "NG"
        
    return setup_res
```

## Function : Create project schema

```python
# Function : create project schema
def schema_setup(catalog, schema):
    setup_res = "OK"

    try:
        spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog}.{schema}")
    except Exception as e:
        print("Error : ", e)
        setup_res = "NG"
        
    return setup_res
```

## Function : Create project volume

```python
# Function : create project volume
def volume_setup(catalog, schema, volume):
    setup_res = "OK"
    spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.{schema}.{volume}")
    return setup_res
```

## Function : Create initial project main program

```python
#Function : initial_setup
def initial_setup():
    try:
        # initial project setup
        spark.sql(f"DROP CATALOG IF EXISTS {ic.catalog_setup} CASCADE")

        # create project catalog
        initial_res = catalog_setup(ic.catalog_setup)

        # create project schema
        if initial_res == "OK":
            for schema in ic.schema_setup:
                initial_res = schema_setup(ic.catalog_setup, schema)

                # create project volume
                if initial_res == "OK":
                    # filter only bronze scheme
                    if schema == ic.schema_setup[0]:
                        for volume in ic.volume_setup:
                            initial_res = volume_setup(ic.catalog_setup, schema, volume)

                            if initial_res == "OK":
                                print(f"Built volume {volume} : Successful")
                            # volume setup failed
                            elif initial_res == "NG":
                                print(f"Volume {volume} setup : Failed")
                            else:
                                print(f"Unknown error {volume}", initial_res)

                # schema setup failed
                elif initial_res == "NG" and schema == ic.schema_setup[0]: 
                    print(f"Schema {schema} setup : Failed")
                else:
                    print(f"Unknown error {schema}", initial_res)

        # catalog setup failed
        elif initial_res == "NG": 
            print(f"Catalog {ic.catalog_setup} setup : Failed")
        else:
            print(f"Unknown error {ic.catalog_setup}", initial_res)

    except Exception as e:
        print("Error : ", e)
    return initial_res
```

## Execute initial project

```python
# execute function
initial_res = initial_setup()
if initial_res == "OK":
    print("Project setup : successful")
else:
    print("Project setup : failed")
```
