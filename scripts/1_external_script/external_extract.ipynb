## 2. Extract data from external source(Github)

### Function : Source data extraction

```python
# data extract from github function
def source_extract(src_config):
    extract_res = "OK"
    
    # download file
    urllib.request.urlretrieve(src_config["url"], src_config["target"])
    
    # check existing file in volume
    if os.path.exists(src_config["target"]):
        extract_res = "OK"
    else:
        print("File not downloaded")
        extract_res = "NG"

    return extract_res
```

### Function : Data extraction main program

```python
def exec_extract():
    extract_res = "OK"
    # process start
    try :
        for src_config in ic.SOURCE_CONFIG:
            extract_res = source_extract(src_config)
            if extract_res == "OK":
                pass
            else :
                return extract_res

    # Exception error
    except Exception as e:
        print("Error : ", e)

    return extract_res
```

### Execute data extraction

```python
extract_res = exec_extract()
if extract_res == "OK":
    print("Data extract completed")
else:
    print("Data extract failed")
```
