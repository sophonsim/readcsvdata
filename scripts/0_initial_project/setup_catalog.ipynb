# Read csv data from github project setup

## Environment configuration

```python
import setup_config as sc
import urllib.request
```

```python

# initial readcsvdata project
spark.sql(f"DROP CATALOG IF EXISTS {sc.catalog_setup} CASCADE;")

# project environment setup function
def setup_config():
    setup_res = "OK"

    try:
        spark.sql(f"CREATE CATALOG IF NOT EXISTS {sc.catalog_setup}")

        for schema in schema_setup:
            spark.sql(f"CREATE SCHEMA IF NOT EXISTS {sc.catalog_setup}.{schema}")

        for volume in volume_setup:
            spark.sql(f"CREATE VOLUME IF NOT EXISTS {sc.catalog_setup}.{sc.schema_setup[0]}.{volume}")
    except Exception as e:
        print("Error : ", e)
        setup_res = "NG"
        
    return setup_res
```

## Ingestion source data from github

```python

# data ingestion from github function
def source_ingest_volume():
    ingest_res = "OK"
    
    try :

        for src_config in sc.SOURCE_CONFIG:
            print(f"Downloading : {src_config['url']}")
            print(src_config["target"])
            urllib.request.urlretrieve(src_config["url"], src_config["target"])
            
    except Exception as e:
        print("Error : ", e)
        ingest_res = "NG"

    return ingest_res

source_ingest_volume()
```

```python
# execute function
setup_res = setup_config()
if setup_res == "OK":
    source_ingest_res = source_ingest_volume()
else:
    print("Setup failed")
```
